CONCLUSION


```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(18, 10))

# Region-wise analysis
region_sales = df.groupby("Region")["Sales"].sum().sort_values(ascending=False)
region_profit = df.groupby("Region")["Profit"].sum().sort_values(ascending=False)
region_colors = ['skyblue', 'lightgreen', 'lightcoral', 'lightsalmon']
region_sales.plot(kind='bar', ax=axes[0, 0], title='Region-wise Sales', color=region_colors)
region_profit.plot(kind='bar', ax=axes[0, 1], title='Region-wise Profit', color=region_colors)

# Segment-wise analysis
segment_sales = df.groupby("Segment")["Sales"].sum().sort_values(ascending=False)
segment_profit = df.groupby("Segment")["Profit"].sum().sort_values(ascending=False)
segment_sales.plot(kind='bar', ax=axes[0, 2], title='Segment-wise Sales', color='steelblue')
segment_profit.plot(kind='bar', ax=axes[1, 0], title='Segment-wise Profit', color='lightcoral')

# State-wise analysis
top_states = df.groupby("State")[["Sales", "Profit"]].sum().sort_values(by="Sales", ascending=False).head(10)
top_states.plot(kind='bar', y='Sales', ax=axes[1, 1], title='Top 10 States by Sales', color='teal')
top_states.plot(kind='bar', y='Profit', ax=axes[1, 2], title='Top 10 States by Profit', color='tomato')

# Adjusting layout
fig.tight_layout(pad=3.0)

# Show plot
plt.show()
```
